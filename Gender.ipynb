```python

def remove_outliers_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]



# Removing outliers from 'Total Social Media Usage' column
cleaned_df = remove_outliers_iqr(df, ' Income (USD)')

print("Original DataFrame:")
print(df)

print("\nCleaned DataFrame:")
print(cleaned_df)
```

```python
# Box plots for numeric variables across categories
numeric_vars = [' Age', ' Height (cm)', ' Weight (kg)', ' Income (USD)']
for var in numeric_vars:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=' Gender', y=var)
    plt.title(f"Box Plot of {var} Across Gender")
    plt.show()
```

```python
categorical_vars = [' Gender', ' Occupation',' Education Level', ' Marital Status', ' Income (USD)',
       ' Favorite Color']
for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=var)
    plt.title(f"Count Plot of {var}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Pie charts
for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    df[var].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title(f"Pie Chart of {var}")
    plt.ylabel("")
    plt.show()
```
